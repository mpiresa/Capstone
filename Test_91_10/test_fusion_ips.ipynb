```python
import pandas as pd
import numpy as np
from scipy.spatial import cKDTree
```

```python
IPS = pd.read_csv("fr-en-ips-ecoles-ap2022.geocoded.csv", sep=";", low_memory=False)
DPE_conso = pd.read_csv("fusion_dpe_enedis.csv", low_memory=False)
```

```python
DPE_conso.columns = DPE_conso.columns.str.strip().str.lower().str.replace(' ', '_')
IPS.columns = IPS.columns.str.strip().str.lower().str.replace(' ', '_')
IPS.columns = IPS.columns.str.replace('é','e').str.replace('è','e').str.replace('à','a')


print(IPS.columns)
DPE_conso.shape
print(DPE_conso.columns)
```

```python
# Compter le nombre d'écoles par commune
ecoles_par_commune = IPS.groupby('nom_de_la_commune').size().sort_values(ascending=False)

# Afficher les 10 communes avec le plus d'écoles
print(ecoles_par_commune.head(10))

# Si tu veux tout sauvegarder dans un CSV
ecoles_par_commune.to_csv("ecoles_par_commune.csv", header=['nombre_ecoles'])
```
