```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
df_dpe_91_2024 = pd.read_csv("df_dpe_91_2024.csv")
df_conso_91_2024 = pd.read_csv("df_conso_91_2024.geocoded.csv")
```

```python
df_conso_dpe_91_2024 = pd.merge(df_conso_91_2024, df_dpe_91_2024, left_on='result_id',  right_on='identifiant_ban', how="inner")
```

```python
df_conso_dpe_91_2024.shape
```

```python
df_conso_dpe_91_2024.head()
```

```python
column_name = "Consommation annuelle moyenne par logement de l'adresse (MWh)"


# Créer un tracé d'histogramme avec la courbe de densité (KDE)
plt.figure(figsize=(10, 6))
sns.histplot(
    data=df_conso_dpe_91_2024, 
    x=column_name, 
    kde=True, 
    bins=30, # Nombre de barres de l'histogramme, ajustable
    color='skyblue',
    edgecolor='black'
)

# Configuration des titres et labels
plt.title(f'Distribution (Gaussienne) de la colonne : {column_name}', fontsize=16)
plt.xlabel(column_name, fontsize=12)
plt.ylabel('Densité (KDE) / Fréquence (Histogramme)', fontsize=12)
plt.grid(axis='y', alpha=0.5)

# Enregistrer la figure
plt.savefig('gauss_consommation_91_2024.png')
plt.close()
```

```python
column_name = "etiquette_dpe"
# Définir l'ordre souhaité pour les catégories (A à G)
ordre_etiquettes = ['G', 'F', 'E', 'D', 'C', 'B', 'A']

# Créer un tracé de comptage (Count Plot)
plt.figure(figsize=(10, 6))

# Utilisez sns.countplot qui est l'outil standard pour ce type de graphique.
sns.countplot(
    data=df_conso_dpe_91_2024,
    x=column_name,
    color='skyblue',
    edgecolor='black',
    # Utilisation du paramètre 'order' pour forcer l'ordre
    order=ordre_etiquettes 
)

# Configuration des titres et labels
plt.title(f'Distribution des Étiquettes DPE (2024) : {column_name}', fontsize=16)
plt.xlabel("Étiquette DPE", fontsize=12)
plt.ylabel('Fréquence', fontsize=12)
plt.grid(axis='y', alpha=0.5)

# Enregistrer la figure
plt.savefig('gauss_dpe_91_2024.png')
plt.close()
```
