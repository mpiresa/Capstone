```python
import requests
import json
import pandas as pd
```

```python
#URL de l'API pour le DPE depuis 2021
base_url = "https://data.ademe.fr/data-fair/api/v1/datasets/dpe03existant/lines"
```

```python
# Paramètres de la requête pour l'Essonne en 2024
params = {
    'code_departement_ban': '91', # Sélection des logements de l'Essonne
    # DPE réalisés en 2024
    'date_visite_diagnostiqueur_lte': '2024-12-31',
    'date_visite_diagnostiqueur_gte': '2024-01-01'
}

response = requests.get(base_url, params=params)
if response.status_code == 200:
    data = response.json()
```

```python
# Création du DataFrame des DPE de l'Essonne en 2024
df_dpe_91_2024 = pd.DataFrame(data['results'])
df_dpe_91_2024.head()
```
