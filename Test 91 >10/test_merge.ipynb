```python
dpe_existant.columns = dpe_existant.columns.str.strip().str.lower().str.replace(' ', '_')
enedis.columns = enedis.columns.str.strip().str.lower().str.replace(' ', '_')
enedis.columns = enedis.columns.str.replace('é','e').str.replace('è','e').str.replace('à','a')
```

```python
dpe_existant['identifiant_ban'] = dpe_existant['identifiant_ban'].astype(str)
enedis['result_id'] = enedis['result_id'].astype(str)
```

```python
merged = pd.merge(
    dpe_existant,
    enedis,
    left_on='identifiant_ban',
    right_on='result_id',
    how='left'  # garder tous les DPE, même si pas de correspondance Enedis
)

print(merged.shape)
merged.head()
```

```python
total_dpe = merged.shape[0]
nb_matched = merged['result_id'].notna().sum()  # lignes avec correspondance
nb_unmatched = merged['result_id'].isna().sum()  # lignes sans correspondance

taux_correspondance = nb_matched / total_dpe * 100

print(f"Total DPE : {total_dpe}")
print(f"Correspondances trouvées : {nb_matched}")
print(f"Pas de correspondance : {nb_unmatched}")
print(f"Taux de correspondance : {taux_correspondance:.2f}%")
```

c'est normal que le taux de correspondance soit si faible car il n'y a que les batiments existants et qui possède un DPE alors que pour Enedi c'est l'ensemble des batiments qui possèdent plus de 10 logements.

### Problème n°1 : Doublons dans le dataset Enedis géocodé avec la BAN

Dans le dataset Enedis après géocodage avec la Base Adresse Nationale (BAN), certains `result_id` correspondent à plusieurs lignes avec des adresses légèrement différentes (numéro de voie, indice de répétition, etc.).

**Choix actuel :**  
Pour la fusion avec le dataset DPE, nous **conservons toutes les occurrences** de ces `result_id` via un **left join**.

**Conséquence :**  
Cela explique pourquoi le nombre de lignes après fusion est supérieur au nombre initial de lignes dans le dataset DPE : certains DPE correspondent à plusieurs lignes Enedis.


```python
# Récupérer la liste des colonnes
colonnes_merge = merged.columns.tolist()

# Écrire dans un fichier texte, une colonne par ligne
with open('colonnes_merge.txt', 'w', encoding='utf-8') as f:
    for col in colonnes_merge:
        f.write(f"{col}\n")
```

```python
print("Avant suppression :", merged.shape)

# Calculer le pourcentage de valeurs manquantes par colonne
taux_na = merged.isna().mean()  # fraction de NaN par colonne

# Sélectionner les colonnes à garder (moins de 90% de NaN)
colonnes_a_garder = taux_na[taux_na <= 0.9].index

# Filtrer le DataFrame
merged_clean = merged[colonnes_a_garder]

# Afficher la forme après suppression
print("Après suppression :", merged_clean.shape)
```
